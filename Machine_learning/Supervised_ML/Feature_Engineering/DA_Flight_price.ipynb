```python
import pandas as pd
```

```python
df = pd.read_excel(r'./flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#splitzinf dataes into new feature and convert to integer format

df['Date'] = df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
# drop date of journy
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minitue'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
df.head()
```

```python
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Departure_Minitue'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.head()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
df.drop('Arrival_Time', axis=1, inplace=True)
df.head()
```

### Dealing with Categorical Values

```python
df.Total_Stops.unique()
import numpy as np
```

```python
df.Total_Stops.isnull().sum()
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,  '1 stop':1, '2 stop':2, '3 stop':3, '4 stop':4, np.nan:1})
```
