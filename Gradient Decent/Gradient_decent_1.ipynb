### General imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

 ### example 1
 #### $f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# plot the data
plt.plot(x_1, f(x_1))

#add limits
plt.xlim([-3, 3])
plt.ylim(0,8)

plt.xlabel('x_1', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.show()
```

### slope and Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
# plot function side by side
plt.figure(figsize=[15,5])

plt.subplot(1, 2, 1)
plt.xlim([-3, 3])
plt.ylim(0,8)
plt.xlabel('x_1', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

#plot1 f(x)

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

#plot2 derivate of f(x)
plt.subplot(1, 2, 2)

plt.xlim([-2, 3])
plt.ylim(-3,6)
plt.xlabel('x_1', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.grid()
plt.plot(x_1, df(x_1), color='red', linewidth=3)

plt.show()
```

### gradeint decent

```python
new_x = 3
previous_x = 0
learning_rate = 0.1
# for a good precision
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - gradient * learning_rate
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print("Number of times loop run ",n)
        break
    
print("Local maxiam occurs ar : ", new_x)
print("Slope or df(x) value at this point is :", df(new_x))
print("Value of loss/cost at this point is : ",f(new_x))

```

```python
# plot function side by side
plt.figure(figsize=[20,5])


plt.subplot(1, 3, 1)
plt.title(" loss/cost Function")

plt.xlim(-3, 3)
plt.ylim(0,8)
plt.xlabel('x_1', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

#plot1 f(x)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

# the x_list in list from scatter
x_list = np.array(x_list)
plt.scatter(x_list, f(x_list), color='red',alpha=0.8)

#==========================================

#plot2 derivate of f(x)

plt.subplot(1, 3, 2)
plt.title("Slope of the loss/cost Function")
plt.xlim(-2, 3)
plt.ylim(-3,6)
plt.xlabel('x_1', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.grid()
plt.plot(x_1, df(x_1), color='red', linewidth=3, alpha=1 )

slope_list = np.array(slope_list)
plt.scatter(slope_list, df(slope_list), color='blue',alpha=0.8)

#plot2 derivate closeup

plt.subplot(1, 3, 3)
plt.title("Gradient decent closeup")
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.xlabel('x_1', fontsize=16)
plt.ylabel('df(x)', fontsize=16)

plt.grid()

plt.plot(x_1, df(x_1), color='red', linewidth=3, alpha=1 )
plt.scatter(x_list, slope_list, color='blue',alpha=0.8)

plt.show()


plt.show()

```
